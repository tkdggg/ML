```python
# 数据集
import numpy as np
datas=iris_datas.data
target=iris_datas.target
datas=np.c_[datas,target]
data1=datas[0:100,:]
```

```python
# 训练集
t2=data1[0:40,0:5]
t1=data1[60:100,0:5]
t=np.vstack((t2,t1))
y=t[:,-1]
# np.shape(t)
x=t[:,0:4]
# 测试集
tc=data1[40:60,:]
yc=tc[:,-1]
xc=tc[:,0:4]
yc
```

```python
# 开始梯度下降
a=0.6
b0=np.random.rand(4)
b0=b0.reshape(4,1)
loss=1
count=0
l=[]
while loss>=0.00001 and count<=100:
    g=0
    loss=0
    
#     for i in range(80):
#         gnow=(1/(1+np.exp(-1*np.dot(x[i],b0)))-y[i])*x[i]
#         g=g+gnow
    i=np.random.randint(0, 80)
    g=(1/(1+np.exp(-1*np.dot(x[i],b0)))-y[i])*x[i]
    g=g.reshape(4,1)
    b0=b0-a*g
#     计算loss
    for k in range(80):
        eloss=(y[k]-1/(1+np.exp(-1*np.dot(x[k],b0))))**2
        loss=loss+eloss
    
#     eloss=y[k]-1/(1+np.exp(-1*np.dot(x[k],b0)))**2
#     loss=loss+eloss
    count=count+1
    l=l+[loss]
    loss=loss/80
    
print(count)
print(b0)
print(loss)
```

```python
# 三分类    softmax 多分类

# 训练集
d12=datas[0:40,0:5]
d13=datas[60:100,0:5]
d14=datas[100:140,0:5]
t=np.vstack((d12,d13,d14))

l1=np.zeros(120).reshape(40,3)
l1[:,0]=1
l2=np.zeros(120).reshape(40,3)
l2[:,1]=1
l3=np.zeros(120).reshape(40,3)
l3[:,2]=1
y=np.vstack((l1,l2,l3))
x=t[:,0:4]

# 测试集
tc1=datas[40:60,:]
tc2=datas[140:150,:]
tc=np.vstack((tc1,tc2))

l1=np.zeros(30).reshape(10,3)
l1[:,0]=1
l2=np.zeros(30).reshape(10,3)
l2[:,1]=1
l3=np.zeros(30).reshape(10,3)
l3[:,2]=1
yc=np.vstack((l1,l2,l3))
xc=tc[:,0:4]
# print(np.shape(x))
y
```

```python
# 进行梯度下降
a2=0.5
count2=0
loss2=1
l2=[]
w0=np.random.rand(3,4)
while loss2>=0.02 and count2<=400:
    i=np.random.randint(0,120)
    za=np.exp(np.dot(w0,x[i].reshape(4,1)))
    g=za/np.sum(za)-y[i].reshape(3,1)
    g=g*x[i]
    w0=w0-a2*g
    count2=count2+1
#     计算loss
    for k in range(120):
        za=np.exp(np.dot(w0,x[k].reshape(4,1)))
        nowloss=za/np.sum(za)-y[k].reshape(3,1)
        loss2=loss2+nowloss**2
    loss2=np.sum(loss2/(120*3))
    l2=l2+[loss2]
# print(l2)
print(w0)
print(count2)
print(loss2)   
```
